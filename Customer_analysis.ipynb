```python
import pandas as pd
```

```python
df = pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df.head(3)
```

```python
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
lables = ['young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=lables)
```

```python
df[['age', 'age_group']].head(15)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping_days = {
    'Annually': 365,
    'Every 3 Months': 90,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Weekly': 7
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping_days)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(12)
```

```python
df[['discount_applied','promo_code_used']].sample(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop(columns = ['promo_code_used'])
```
